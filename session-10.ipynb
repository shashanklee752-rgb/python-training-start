```python
import numpy as np
import sys
import time

# 1. Create a large list and an equivalent NumPy array (1,000,000 elements)
size = 1000000
python_list = list(range(size))
numpy_array = np.arange(size)

# 2. Estimate memory usage
# For the list, we sum the size of the list object plus the size of every integer object within it
list_mem = sys.getsizeof(python_list) + sum(sys.getsizeof(i) for i in python_list)
# For NumPy, we use the nbytes attribute for the contiguous memory block
numpy_mem = numpy_array.nbytes

print(f"Python list memory: ~{list_mem} bytes; NumPy array memory: {numpy_mem} bytes")

# 3. Measure summation speed
# Timing Python's built-in sum on the list
start_list = time.time()
sum(python_list)
time_list = time.time() - start_list

# Timing NumPy's optimized sum on the array
start_numpy = time.time()
numpy_array.sum()
time_numpy = time.time() - start_numpy

print(f"Time using Python list: ~{time_list:.6f} seconds")
print(f"Time using NumPy array: ~{time_numpy:.6f} seconds")
```
